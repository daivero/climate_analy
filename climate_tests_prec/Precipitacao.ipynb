```python
import pandas as pd
```

```python
precipitacao=pd.read_excel('https://portal.inmet.gov.br/uploads/normais/Normal-Climatologica-PREC.xlsx',skiprows=2)
precipitacao
```

```python
import numpy as np
precipitacao[['UF', 'Janeiro', 'Fevereiro', 'Março',
       'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro',
       'Novembro', 'Dezembro']].replace('-', np.nan)
```

```python
prec=precipitacao[['UF', 'Janeiro', 'Fevereiro', 'Março',
       'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro',
       'Novembro', 'Dezembro']].replace('-', np.nan).groupby("UF").mean()#.mean(axis=0)
```

```python
prec
```

```python
for idx, row in prec.iterrows():
    average=row.mean()
    prec.at[idx,'prec']=average
```

```python
prec['prec'].to_dict()
```

```python
prec_dict=prec['prec'].to_dict()
```

```python
import geopandas as gpd

# Make sure the path to the shapefile is correct
shapefile_path = 'BRA_adm1.shp'
gdf_brazil = gpd.read_file(shapefile_path)
```

```python
for idx, row in gdf_brazil.iterrows():
    state=row['HASC_1'].replace('BR.','')
    if state in prec_dict:
        prec_for_state=prec_dict[state]
        gdf_brazil.at[idx,'prec']=prec_for_state
```

```python
gdf_brazil.plot(column='prec', legend=True)
```
